# 📌 .at[]
### .at[] is used to access a single value in a DataFrame by label (row index and column name). It's faster than .loc[] when accessing individual elements.

```python
data.at[5, 'Title']
```

```python
data.iloc[1: 150:]
```

# (Using .at[])
# Retrieve the 'Country' of the entry at row index 150 and then check if it matches 'United States'. If it matches, replace it with 'USA'.

```python
# DataFrame.at[row_label, column_label]
```

```python
if data.iloc[147, 5] == 'United States':
    print("The value exists at that position.")
else:
    print("The value does not exist at that position.")
```

```python
if data.at[147, 'Country'] == "United States":
    data.at[147, 'Country'] = "USA"
```

```python
print(data.at[147, 'Country'])
```

# (Using .at[])
# Update the 'Category' of the entry at row index 300 to 'Documentary' only if its 'Type' is 'Movie'.

```python
try:
    row_data = data.iloc[300]
    print("Row 300 exists. Here's the data:\n", row_data)
except IndexError:
    print("Row 300 does not exist in the DataFrame.")
```

```python
if data.at[300, 'Category'] == 'Movie':
    data.at[300, 'Category'] = 'Documentary'
```

```python
data.at[300, 'Category']
```

# Check if the value at row index 500 and column index 4 is missing (NaN). If it is, replace it with the text 'Unknown'.

```python
if pd.isna(data.iat[500, 4]):  # Check if the value is NaN
    data.iat[500, 4] = 'Unknown'  # Replace it with 'Unknown'
```

```python

print(data.iat[500, 4])  # This will print the value at row index 500, column index 4
```
